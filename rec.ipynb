```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

df=pd.read_csv(r"C:\Users\vamsv\Downloads\metafinals.csv")
df['interests']=[[i] for i in df['interests']]
df['course']=[[i] for i in df['course']]
df
```

```python
tfidf = TfidfVectorizer(tokenizer=lambda x: x, lowercase=False)

projects_tfidf = tfidf.fit_transform(df['interests'])
course_tfidf = tfidf.fit_transform(df['course'])

project_similarity = linear_kernel(projects_tfidf, projects_tfidf)
course_similarity = linear_kernel(course_tfidf, course_tfidf)
combined_similarity=project_similarity+course_similarity/2
```

```python
def get_recommendations(student_name, num_recommendations=3):
    student_index = df[df['first_name'] == student_name].index[0]
    similar_students_indices = combined_similarity[student_index].argsort()[::-1][1:]
    recommended_students = df.iloc[similar_students_indices[:num_recommendations]]
    return recommended_students[['first_name', 'last_name', 'interests']]


student_name = 'Donalt'

recommendations = get_recommendations(student_name)
print(f"Recommended students for {student_name}:")
print(recommendations)
```
